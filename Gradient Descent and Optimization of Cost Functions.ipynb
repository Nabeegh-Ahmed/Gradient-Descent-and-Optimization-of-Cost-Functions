```python
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm # color map
import random
```

## Cost Function 1: $$f(x)=x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1

def df(x):
    return 2*x + 1
```

```python
testData = np.linspace(start=-5, stop=5, num = 500)
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title('Cost Function')
plt.plot(testData, f(testData))
plt.xlabel('Data')
plt.ylabel('f(Data)')
plt.xlim(-3, 3)
plt.ylim(0, 13)

plt.subplot(1, 2, 2)
plt.title('Slope of Cost Function')
plt.plot(testData, df(testData))
plt.xlabel('Data')
plt.ylabel('df(Data)')
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.show()
```

### Gradient Descent Function

```python
def gradientDescent(derivativeFunction, _coefficient=random.randint(-5, 5), _alpha=0.1, _precision=0.00001, max_steps=500):
    coefficient =  _coefficient
    alpha = _alpha # Learning Rate
    delta = derivativeFunction(coefficient)
    precision = _precision

    ChangeInCoefficient = [coefficient]
    ChangeInSlope = [delta]

    for n in range(max_steps):
        lastCoefficient = coefficient
        coefficient = coefficient - (alpha * delta)
        delta = derivativeFunction(coefficient)
        ChangeInCoefficient.append(coefficient)
        ChangeInSlope.append(delta)

        if abs(coefficient - lastCoefficient) < precision:
            break
    
    return coefficient, ChangeInCoefficient, ChangeInSlope
```

```python
LocalMinimum, ChangeInCoefficient, ChangeInSlope = gradientDescent(df)

plt.figure(figsize=[15, 5])

plt.subplot(1, 3, 1)
plt.title('Cost Function')

plt.xlabel('Data')
plt.ylabel('f(Data)')
plt.xlim(-3, 3)
plt.ylim(0, 13)

plt.plot(testData, f(testData))

ChangeInCoefficientValues = np.array(ChangeInCoefficient)
plt.scatter(ChangeInCoefficient, f(ChangeInCoefficientValues), color='red', alpha=0.6, s=100)

plt.subplot(1, 3, 2)
plt.title('Slope of Cost Function')
plt.xlabel('Data')
plt.ylabel('df(Data)')
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(testData, df(testData), color='skyblue')
plt.scatter(ChangeInCoefficient, ChangeInSlope, color='red', alpha=0.6, s=100)

plt.subplot(1, 3, 3)
plt.title('Slope of Cost Function (Close UP)')
plt.xlabel('Data')
plt.ylabel('df(Data)')
plt.grid()
plt.xlim(-0.53, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(testData, df(testData), color='skyblue')
plt.scatter(ChangeInCoefficient, ChangeInSlope, color='red', alpha=0.6, s=100)

plt.show()
```

## Cost Function 2: $$f(x)=x^4 - 4x^2 + 5$$

```python
testData_2 = np.linspace(-2, 2, 1000)

# Challenge: Write the g(x) function and the dg(x) function in Python?
def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title('Cost Function')
plt.plot(testData_2, g(testData_2))
plt.xlabel('Data')
plt.ylabel('g(Data)')
plt.xlim(-2, 2)
plt.ylim(1, 6)

plt.subplot(1, 2, 2)
plt.title('Slope of Cost Function')
plt.plot(testData_2, dg(testData_2))
plt.xlabel('Data')
plt.ylabel('dg(Data)')
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.show()
```

```python
LocalMinimum_g, ChangeInCoefficient_g, ChangeInSlope_g = gradientDescent(dg, 0.5) 
# Please note that depending on the Learning Rate(alpha), the result might be extremely large and handeling of this large result
# might not be possible on certain machines. So change the maxing steps or change the learning Rate

plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title('Cost Function')
plt.xlabel('Data')
plt.ylabel('g(Data)')
plt.xlim(-2, 2)
plt.ylim(1, 6)

plt.plot(testData_2, g(testData_2))
plt.scatter(ChangeInCoefficient_g, g(np.array(ChangeInCoefficient_g)), color='red', alpha=0.6, s=100)

plt.subplot(1, 2, 2)
plt.title('Slope of Cost Function')
plt.xlabel('Data')
plt.ylabel('dg(Data)')
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.plot(testData_2, dg(testData_2))

plt.plot(testData_2, dg(testData_2), color='skyblue')
plt.scatter(ChangeInCoefficient_g, ChangeInSlope_g, color='red', alpha=0.6, s=100)

plt.show()
```

## Cost Function 3:  $$h(x) = x^5 - 2x^4 + 2$$

```python
testData_3 = np.linspace(start=-2.5, stop=2.5, num=1000)

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 - 8*x**3
```

```python
LocalMinimum_h, ChangeInCoefficient_h, ChangeInSlope_h = gradientDescent(dh, -0.2, max_steps=19)
# Please note that depending on the Learning Rate(alpha), the result might be extremely large and handeling of this large result
# might not be possible on certain machines. So change the maxing steps or change the learning Rate

plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title('Cost Function')
plt.xlabel('Data')
plt.ylabel('h(Data)')
plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)

plt.plot(testData_3, h(testData_3))
plt.scatter(ChangeInCoefficient_h, h(np.array(ChangeInCoefficient_h)), color='red', alpha=0.6, s=100)

plt.subplot(1, 2, 2)
plt.title('Slope of Cost Function')
plt.xlabel('Data')
plt.ylabel('dh(Data)')
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)

plt.plot(testData_3, dh(testData_3))
plt.scatter(ChangeInCoefficient_h, ChangeInSlope_h, color='red', alpha=0.6, s=100)

plt.show()
```

## Change in Behavior of Gradient Descent Values Based on Variation of Learning Rate

```python
# For this experiment, I am using f(x) (use g(x) or h(x) if your machine is a good one)
maxSteps = 100 # Same number of maxSteps for will help in plotting
lowAlpha  = gradientDescent(df, _alpha=0.0005, _coefficient=3, max_steps=maxSteps)
midAlpha  = gradientDescent(df, _alpha=0.005,  _coefficient=3, max_steps=maxSteps)
HighAlpha = gradientDescent(df, _alpha=0.05,   _coefficient=3, max_steps=maxSteps)

numberOfIterations= list(range(0, maxSteps+1))

plt.figure(figsize=[20, 10])

plt.xlim(0, maxSteps)
plt.ylim(0, 50)

plt.title('Effect of the learning rate', fontsize=17)
plt.xlabel('Number of iterations', fontsize=16)
plt.ylabel('Cost', fontsize=16)

plt.plot(numberOfIterations, f(np.array(lowAlpha[1])), color='lightgreen', linewidth=5)
plt.plot(numberOfIterations, f(np.array(midAlpha[1])), color='steelblue', linewidth=5)
plt.plot(numberOfIterations, f(np.array(HighAlpha[1])), color='hotpink', linewidth=5)

plt.show()
```
